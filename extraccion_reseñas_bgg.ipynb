```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Configuración más detallada de las opciones de Chrome
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--disable-gpu')

# Configurar la ruta del driver
service = Service('/usr/bin/chromedriver')
```

```python
def get_comments(delay=2, pages=1):
    """
    Obtiene reseñas del juego, se puede obtener diferentes páginas de comentarios cambiando el parametro "page".
    """
    # Configurar Selenium con WebDriver Manager
    driver = webdriver.Chrome(options=chrome_options)
    try:
      comment_list = []
      for page in range(pages):
          url = f"https://boardgamegeek.com/boardgame/199561/sagrada/ratings?pageid={page+1}&status=own&comment=1&rated=1"
          driver.get(url)
          print(f"⏳ Esperando {delay} segundos para que carguen los datos...")
          time.sleep(delay)  # Esperar para permitir que el JavaScript cargue los datos dinámicos

          # Obtiene los comentarios de la página
          summary = driver.find_element(By.CLASS_NAME, "summary")

          if summary:
            # Obtiene cada comentario
            elementos = summary.find_elements(By.CLASS_NAME, "summary-item")
            for elemento in elementos:

              #Extrae el texto del comentario
              comment = elemento.find_element(By.CLASS_NAME, "mb-0")

              # Guarda el texto del comentario en la lista
              comment_list.append(comment.text) 

          else:
              print("⚠ No se encontraron comentarios en la página.")

    finally:
        driver.quit()  # Cerrar el navegador
        return comment_list
```

```python
#Guardamos los comentarios de las primeras 50 páginas
comments = get_comments(pages=50)
```

```python
#Guardamos los comentarios en un archivo de texto
with open("./datos/informacion/sagrada_comments.txt", "w", encoding="utf-8") as file:
    for comment in comments:
        file.write(comment + "\n\n" + ";" + "\n\n")
print("Comentarios guardados en sagrada_comments.txt")
```
